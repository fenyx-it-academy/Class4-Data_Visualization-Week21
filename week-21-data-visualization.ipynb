```python
data = pd.read_csv('../input/weather-dataset-rattle-package/weatherAUS.csv')

df = data.copy()
df
```

```python
df.describe().T
```

```python
df.info()
```

```python
unique_val = []
for i in df.columns:
    u = df[i].nunique()
    unique_val.append(u)
    
pd.DataFrame({"No. of unique values": unique_val}, index=df.columns)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['RainTomorrow', 'RainToday'])
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df['Year'] = df['Date'].dt.year.astype('int16')
df['Month'] = df['Date'].dt.month.astype('int16')
df['Year_Month']=df['Date'].array.strftime('%Y-%m') 
df['DiffTemp']=df['MaxTemp']-df['MinTemp']
df.head()
```

# 1- MissingNo visualizations

```python
miss = df.isnull().sum().sort_values(ascending = False).head(5)
miss_per = (miss/len(df))*100
pd.DataFrame({'No. missing values': miss, '% of missing data': miss_per.values})
```

```python
msno.matrix(df);
```

```python
msno.bar(df);
```

# 2-Frequency of categorical variables with histogram

```python
cols_obj=[
 'Location',
 'WindGustDir',
 'WindDir9am',
 'WindDir3pm',
 'RainToday',
 'RainTomorrow'
 ]
```

```python
for i in cols_obj[:-1]:
   
    fig= px.histogram(df, x=i)
    
    fig.show()
```

# 3-Distribution of numerical variables with histogram and boxplot/violin

```python
df = df.dropna(subset=['RainTomorrow', 'RainToday'])
```

```python
cols=[
 'MinTemp',
 'MaxTemp',
 'Rainfall',
 'Evaporation',
 'Sunshine',
 'WindGustSpeed',
 'WindSpeed9am',
 'WindSpeed3pm',
 'Humidity9am',
 'Humidity3pm',
 'Pressure9am',
 'Pressure3pm',
 'Cloud9am',
 'Cloud3pm',
 'Temp9am',
 'Temp3pm',
 'DiffTemp',
 'Year',
 'Month']
```

```python
for i in cols[:-1]:
    
    fig= px.histogram(df, x=i,y="Location",color="RainToday")
    
    fig.show()
```

# 4-Heatmap for correlation

```python
msno.heatmap(df);
```

# 5-Scatter matrix and/or OLS Scatter Plot for numerical variables

```python
cols_sm=[
 'MinTemp',
 'MaxTemp',
 'Rainfall',
 'Evaporation',
 'Sunshine',
 'WindGustSpeed',
 'WindSpeed9am',
 'WindSpeed3pm',
 'Humidity9am',
 'Humidity3pm',
 'Pressure9am',
 'Pressure3pm',
 'Cloud9am',
 'Cloud3pm',
 'Temp9am',
 'Temp3pm',
 'DiffTemp',
 'Year',
 'Month']
```

```python
df1 = df.groupby(['Year_Month','RainToday'])[cols_sm].agg('mean').reset_index()
```

```python
for i in cols_sm[:-1]:
    
    fig = px.scatter(df1, x=i,color="RainToday"
                    
                     )
    fig.show()
```

# 6-Time series (preferably monthly) line chart for numerical variables

[](http://)

```python
cols_time= ['MinTemp', 'MaxTemp', 'DiffTemp','Rainfall', 'Evaporation',
       'Sunshine','WindGustSpeed','WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm']
```

```python
df1 = df.groupby(['Year_Month','RainToday'])[cols].agg('mean').reset_index()
```

```python
for i in cols_time:
    
    fig = px.line(df1, x='Year_Month', y=i, 
                     color='RainToday',
                 )
    fig.show()
```
