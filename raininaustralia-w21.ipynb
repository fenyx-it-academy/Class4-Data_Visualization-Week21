# Frequency of categorical variables with histogram

```python
df.dtypes
```

```python
for i in df.columns:
    if df.dtypes[i]=='O':
        fig = px.histogram(df,x=i)
        fig.show()
```

# Distribution of numerical variables with histogram and boxplot/violin

```python
numerical_var = [i for i in df.columns if df[i].dtype!='O']
```

```python
for i in numerical_var:
    df[i].replace(to_replace=np.nan, value=df[i].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    if df.dtypes[i]=='O':
        df[i].replace(to_replace=np.nan, value='other', inplace=True)
```

```python
df.isnull().sum()
```

```python
for i in df[numerical_var]:
    fig=px.histogram(df, x=i,color='RainToday')
    fig.show()
```

```python
for i in df[numerical_var]:
    fig=px.violin(df, y=i, box=True)
    fig.show()
```

```python
fig = plt.figure(figsize=(18,11))
number_of_fig = 1
for i in df[numerical_var]:   
    plt.subplot(6, 2, number_of_fig)
    sns.distplot(df[i])
    number_of_fig +=1
    
plt.tight_layout()
plt.show()
```

# Heatmap for correlation

```python
df['Date'] =pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df.head()
```

```python
plt.figure(figsize=(14,12))
ax = sns.heatmap(df.corr(), square=True, annot=True, fmt='.1f')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)          
plt.show()
```

```python
df.corr()
```

# Scatter matrix and/or OLS Scatter Plot for numerical variables
# Additionally RainToday and RainTomorrow variables as color dimension

```python
for i in numerical_var:
    fig=px.scatter(df, y='Rainfall', x=i, color='RainToday',trendline='ols')
    fig.show()
```

# Time series (preferably monthly) line chart for numerical variables
# RainToday and RainTomorrow variables as 3rd variable with color dimension
# Tip: Use groupby method to group by month/year and calculate mean values

```python
for i in numerical_var:
    fig = px.line(df, x="Month", y=i, color='RainToday')
    fig.show()
```
